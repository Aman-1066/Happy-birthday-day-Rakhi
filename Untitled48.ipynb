```python
# Zero-Shot Prompt Example

prompt = "What is 2 + 3?"

print("Zero-Shot Prompt:")
print(prompt)

# Direct answer without steps
answer = 5
print("AI Answer:", answer)
```

```python
# Chain of Thought Prompt Example

prompt = """
Find 30% of 20 and then subtract 4.
Please explain step by step.
"""

print("Chain of Thought Prompt:")
print(prompt)

# Step-by-step reasoning
step1 = 20 * 30 / 100
print("Step 1: 30% of 20 =", step1)

step2 = step1 - 4
print("Step 2: After subtracting 4 =", step2)

print("Final Answer:", step2)
```

```python

print("----- Zero-Shot -----")
print("Question: 10 + 5")
print("Answer: 15")

print("\n----- Chain of Thought -----")
print("Question: 10 + 5")
print("Step 1: Add 10 and 5")
print("Final Answer: 15")
```
